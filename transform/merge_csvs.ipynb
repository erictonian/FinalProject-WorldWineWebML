```python
# import the dependencies
import pandas as pd
import numpy as np
```

```python
# import the two CSVs
wine_data_original = "../data/winemag-data-130k-v2.csv"
new_scraped_wine = "test_output.csv"
```

```python
# read the csv's into a dataframe
wine_data_df = pd.read_csv(wine_data_original)
new_scraped_df = pd.read_csv(new_scraped_wine)
```

```python
# view the head of the CSVs
wine_data_df.head()
```

```python
# view the heado of the new data CSV
new_scraped_df.head()
```

```python
# combine the two dataframes without altering any data
merged_df = pd.concat([wine_data_df, new_scraped_df], axis = 0)
```

```python
# view the new merged file
merged_df.head()
```

```python
# drop the unnamed column
merged_df = merged_df.drop(columns=["Unnamed: 0"])
```

```python
# export the files to a .csv file
merged_df.to_csv("../data/wine_data_190K.csv", index = False)
```
